```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("amazon.csv",encoding='ISO-8859-1')
df["date"]=pd.to_datetime(df["date"])
```

```python
df=df.drop_duplicates()
```

```python
df["month"]=df["month"].replace(['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho', 'Julho',
       'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],["January","February","March","April","May","June","July","August","September","October","November","December"])
```

No.of Forest Fire per Year

```python
df.groupby("year")["number"].sum().sort_values(ascending=False)
```

```python
df_year=df.groupby("year")["number"].sum().reset_index()
df_year=df_year.sort_values(by="number",ascending=False)
plt.figure(figsize=(15,10))
sns.barplot(x="year",y="number",data=df_year).set(title="No.of Forest Fire per Year")
```

No.of Forest Fire per Month

```python
df_month=df.groupby("month")["number"].sum().sort_values(ascending=False)
df_month=df_month.reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="month",y="number",data=df_month).set(title="No.of Forest Fire per Month")
```

No.of Forest Fire per State

```python
df_state=df.groupby("state")["number"].sum().reset_index()

plt.figure(figsize=(15,10))
sns.barplot(x="number",y="state",data=df_state).set(title="No.of Forest Fire per State")
```
